```python
afthrs_count = lrn_count + ln_count
afthrs_count
```

```python
# Generate a bar plot showing the total number of pick up points for yellow cabs in Manhattan
# in December of 2019 
# Mathplotlib - pyplot.

pup_count = [morn_count, day_count, eve_count, afthrs_count]
x_axis = np.arange(4)

#x_axis = np.arange(len(users))

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Morning", "Day", "Evening", "Late Night"])


plt.bar(x_axis, pup_count, color='r', alpha=0.5, align="center")
plt.xticks()
plt.title("Yellow Cabs in Manhattan - December 2019")
plt.ylabel("Number of Pickups")
plt.xlabel("Time of Day")
plt.show()
```

```python
Dec_2019 = [morn_count, day_count, eve_count, afthrs_count]
June_2019 = [june_morn_count, june_day_count, june_eve_count, june_afthrs_count]
index = ["5am-11am", "11am-5pm", "5pm-11pm", "11pm-5am"]
df = pd.DataFrame({'Dec_2019': Dec_2019,
                   'June_2019': June_2019}, index=index)
ax = df.plot.bar(rot=0)
plt.title("Yellow Cabs in Manhattan")
plt.ylabel("Number of Pickups")
plt.xlabel("Time of Day")
```
